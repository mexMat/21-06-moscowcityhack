```python
import pandas as pd
import requests
import json
import re

from IPython.display import clear_output
from bs4 import BeautifulSoup
from time import sleep
```

```python
HOST = 'https://netology.ru'
MAIN_LIST = '/navigation?direction=development%2Cdata-science%2Cdesign'

HEADERS = requests.utils.default_headers()
HEADERS.update({'user-agent':'Mozilla/5.0 (X11; Linux x86_64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/91.0.4472.101 Safari/537.36',
                'accept':'*/*'})
```

```python
def get_html_page(url, n_attempts=5, t_sleep=1):
    
    while n_attempts:
        try:
            r = requests.get(url, headers=HEADERS)
            r.encoding = 'utf-8'
            if r.status_code != 200:
                return None
            
            html = r.text
            r.close()
            break
        
        except requests.exceptions.RequestException as e:
            n_attempts -= 1
            print(f'Sleep for {t_sleep} sec; {n_attempts} left')
            
            sleep(t_sleep)
            continue
            
    return html
```

```python
def cleaner(string):
    string = re.sub('\n+|\t+|\r+', '', string)
    while string[0] in ' ,.':
        string = string[1:]
    
    while string[-1] == ' ':
        string = string[:-1]
    
    string = re.sub('\ +', ' ', string)
    string = re.sub('\. ', '.', string)
    string = re.sub(' \.', '.', string)
    string = re.sub(', ', ',', string)
    string = re.sub(' ,', ',', string)
    
    return string
```

```python
def get_course_info(url, url_for_user):
    try:
        raw_json = get_html_page(url)
        info_json = json.loads(raw_json)
        l = list()
    
        for i in info_json['blocks']:
            name = re.sub('Блок', '', i['title'])
            l.append(name)
        return {'text' : l,
                'url_for_user' : url_for_user}

    except TypeError as e:
        return {'text' : '',
                'url_for_user' : ''}
```

```python
def main_parse():
    raw_list = get_html_page('https://netology.ru/backend/api/programs')
    list_professions = json.loads(raw_list)
    main_list = list()
    
    for json_obj in list_professions:
        course_name  = json_obj['name']
        url_for_user = 'https://' + json_obj['url'][2:]
        
        
        if len(url_for_user.split('/programs/')) > 1:
            try:
                l = list()
                m = list()

                prog_name    = url_for_user.split('/programs/')[1]
                url_for_parsing = f'https://netology.ru/backend/api/programs/{prog_name}'


                print(f'{course_name}\n{url_for_user}\n{url_for_parsing}\n' + '-'*55)

                raw_json = get_html_page(url_for_parsing)
                info_json = json.loads(raw_json)

                for i in info_json['blocks']:
                    name = re.sub('Блок', '', i['title'])
                    l.append(name)
                
                if 'skills' in info_json:
                    for i in info_json['skills']:
                        m.append(i)

                main_list.append({
                                    'url_for_user' : url_for_user,
                                    'blocks'       : l,
                                    'skills'       : m
                                 })
            except TypeError as e:
                continue
        
    return main_list
```

```python
FRAME_NETOLOGY = main_parse()
```

```python
FRAME_NETOLOGY = pd.DataFrame(FRAME_NETOLOGY)
```

```python
def get_list_of_idxs():

    page = 1
    list_idxs = list()
    c = 0
    
    while True:
    
        raw_json = get_html_page(f'https://stepik.org:443/api/courses?page={page}')
        json_list = json.loads(raw_json)
        
        is_next = json_list['meta']['has_next']
        
        for item in json_list['courses']:
            idx = item['id']
            c += 1

            list_idxs.append(idx)
        if page % 10 == 0:
            print(f'PAGE: {page}, NUM: {c}')            
        
        page += 1
        
        if not is_next:
            break
            
    return list_idxs
```

```python
def parse_stepik(list_of_idxs):
    main_list = list()
    c = 0
    
    for idx in list_of_idxs:
        
        url_to_course = f'https://stepik.org/course/{idx}/promo'
        
        html = get_html_page(url_to_course)
        soup = BeautifulSoup(html, 'html.parser')
        text = soup.find('body').text
        
        main_list.append({
                            'id'   : idx,
                            'url'  : url_to_course,
                            'text' : cleaner(text)
                         })
        
        print(f'{c}/12234+')
        c += 1
        
    return main_list        
```

```python
def parse():
    HOST = 'https://ru.coursera.org'
    html = get_html_page(HOST + '/browse')

    soup = BeautifulSoup(html, 'html.parser')
    professions = soup.find_all('a', class_='explore-domains-card nostyle')
    main_list = list()
    
    for profession in professions:
        prof_name = profession['aria-label']
        prof_link = profession['href']
        prof_link = HOST + prof_link
        
        html = get_html_page(prof_link)
        if html:

            soup = BeautifulSoup(html, 'html.parser')


            course_lists = soup.find_all('div', class_='slick-list')

            for item in course_lists:
                courses_group = item.find_all('div', class_='slick-slide')

                for course in courses_group:
                    link_to_course = course.find('a')

                    if link_to_course:
                        link_to_course = HOST + link_to_course['href']
                        print(link_to_course)
                        print('-'*70)

                        course_page = get_html_page(link_to_course)
                        if course_page:
                            soup = BeautifulSoup(course_page, 'html.parser')
                            key_space = soup.find('div', class_='Skills')
                            skill_list = list()

                            if key_space:
                                keys = key_space.find_all('span')
                                for key in keys:
                                    try:

                                        skill_list.append(key['title'].lower())
                                        main_list.append({
                                                    'profession_name'    : prof_name,
                                                    'link_to_profession' : prof_link,
                                                    'link_to_course'     : link_to_course,
                                                    'skills'             : skill_list
                                                })

                                    except KeyError as e:
                                        continue

    return main_list
```
